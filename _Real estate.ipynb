```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(df.drop(['Y house price of unit area'],axis=1), 
                                                    df['Y house price of unit area'], test_size=0.10, 
                                                    random_state=10)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
accuracy=model.score(X_test,y_test)
print(accuracy*100,'%')
```
